```python
import cv2
import numpy as np
from insightface.app import FaceAnalysis
from google.colab.patches import cv2_imshow  # Required for displaying images in Colab

app = FaceAnalysis()
app.prepare(ctx_id=0)  # Use GPU (ctx_id=0) or CPU (ctx_id=-1)

# Load image
image = cv2.imread("sri divya.jpg")

# Perform face detection
faces = app.get(image)

# Draw bounding boxes and landmarks
for face in faces:
    x, y, x2, y2 = face.bbox.astype("int")
    cv2.rectangle(image, (x, y), (x2, y2), (0, 255, 0), 2)

    for (px, py) in face.kps.astype("int"):
        cv2.circle(image, (px, py), 2, (0, 0, 255), -1)

# Show output in Google Colab
cv2_imshow(image)
```
